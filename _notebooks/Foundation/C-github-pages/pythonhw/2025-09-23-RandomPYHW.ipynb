```python
import random

def open_loot_boxes(rare_chance):
    rare_count = 0
    common_count = 0

    print(f"\nOpening 10 loot boxes with a {rare_chance}% rare chance...\n")

    for i in range(10):
        roll = random.randint(1, 100)
        if roll <= rare_chance:
            print(f"Loot Box {i+1}: ğŸ RARE item!")
            rare_count += 1
        else:
            print(f"Loot Box {i+1}: ğŸª™ Common item.")
            common_count += 1

    print("\n--- RESULTS ---")
    print(f"Rare items: {rare_count}")
    print(f"Common items: {common_count}")

# "Slider" simulated with input
chance = int(input("Enter rare item chance (10 to 90): "))

# Make sure it's in range
if 10 <= chance <= 90:
    open_loot_boxes(chance)
else:
    print("Please enter a value between 10 and 90.")
```
