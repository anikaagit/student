```python
%%js

//  Random Password Maker
const chars =
  "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789";

function generatePassword() {
  const length = parseInt(
    prompt("Enter password length (6‚Äì20):", "10")
  );

  if (isNaN(length) || length < 6 || length > 20) {
    alert("‚ùå Please enter a number between 6 and 20.");
    return;
  }

  let password = "";
  for (let i = 0; i < length; i++) {
    const randomIndex = Math.floor(Math.random() * chars.length);
    password += chars[randomIndex];
  }

  // Display result
  const output =
    document.getElementById("password-output") ||
    (() => {
      const p = document.createElement("p");
      p.id = "password-output";
      p.className = "password";
      document.body.appendChild(p);
      return p;
    })();
  output.textContent = `Generated Password: ${password}`;
}

// Create button if not already in HTML
if (!document.getElementById("password-btn")) {
  const btn = document.createElement("button");
  btn.id = "password-btn";
  btn.textContent = "üîÅ Generate Password";
  btn.onclick = generatePassword;
  document.body.appendChild(btn);
}
```

```python
// Number Guessing Game
let randomNumber = Math.floor(Math.random() * 50) + 1;
let guessCount = 0;

function guessNumber() {
  const userGuess = parseInt(prompt("Guess a number between 1 and 50:"));

  if (isNaN(userGuess) || userGuess < 1 || userGuess > 50) {
    alert(" Please enter a valid number between 1 and 50.");
    return;
  }

  guessCount++;

  if (userGuess === randomNumber) {
    alert(` Correct! You guessed it in ${guessCount} tries!`);
    resetGame();
  } else if (userGuess < randomNumber) {
    alert(" Too low! Try again.");
  } else {
    alert(" Too high! Try again.");
  }
}

function resetGame() {
  if (confirm("Play again?")) {
    randomNumber = Math.floor(Math.random() * 50) + 1;
    guessCount = 0;
  }
}

// Create button if not already in HTML
if (!document.getElementById("guess-btn")) {
  const btn = document.createElement("button");
  btn.id = "guess-btn";
  btn.textContent = " Start Guessing Game";
  btn.onclick = guessNumber;
  document.body.appendChild(btn);
}
```
